```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
df= pd.read_csv('D:\\dsa\\data2018.csv')
x = df[["Add1(in Thousands)", "Add2(in Thousands)", "Add3(in Thousands)", "Add4(in Thousands)", 'Add5(in Thousands)']]
y = df["Profit(in Thousands)"]

x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1, 1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
data= pd.read_csv('D:\\dsa\\data2018.csv')
data = pd.DataFrame(data={"Predicted Profit": ypred.flatten()})
print(data.head())
```
