```python
# Here we use  anova to test the following hypothesis

#H0: All Zones generate the same amount of sales on average.
#H1: Atleast average sales on one zone is diffent
```

```python
import scipy.stats as stat
```

```python
F_value,p_value = stat.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
```

```python
print("F statistics value : ",round(F_value,5))
print("p value : ",round(p_value,5))
```

```python
if p_value<0.05:
  print("Reject H0 at 5% level of significance")
else:
  print("Fail to reject H0 at 5% level of significance")
```

```python
# the p value is less than 0.05 so reject the null hupothesis.
```
