```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('D:\\dsa\\StudentsPerformance.csv')
df.shape
```

```python
plt.subplot(221)

df['test preparation course'].value_counts().plot(kind='bar', title='Test preparation course')

plt.xticks(rotation=0)
plt.subplot(222)

df['gender'].value_counts().plot(kind='bar', title='Gender of students', figsize=(16,9))

plt.xticks(rotation=0)

plt.show()
```

```python
# 2 The students' parental level of education
```

```python
df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().T.plot(figsize=(10,8))
```

```python
plt.rcParams['figure.figsize'] = (15, 9)
sns.countplot(x = 'parental level of education', data = df, hue = 'test preparation course', palette = 'dark')
plt.title('Parental Education vs Test Preparation Course', fontweight = 30, fontsize = 20)
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=0)
plt.show()
```
